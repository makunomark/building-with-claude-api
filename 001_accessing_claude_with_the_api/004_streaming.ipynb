```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1024,
        "messages": messages,
        "temperature": temperature,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = []

add_user_message(messages, "write a one sentence joke about computers")

stream = client.messages.create(
    model=model,
    max_tokens=1024,
    messages=messages,
    stream=True,
)

for chunk in stream:
    print(chunk)
```

```python
messages = []

add_user_message(messages, "Write a one sentence poem about the sea.")

with client.messages.stream(
    model=model,
    max_tokens=1024,
    messages=messages,
) as stream:
    for test in stream.text_stream:
        print(test, end="")

    # getting the final message object
    print(stream.get_final_message())
```
